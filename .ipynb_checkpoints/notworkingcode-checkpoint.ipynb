````python
from scipy.spatial.transform import Rotation
import matplotlib.animation as mpla
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
import scipy
````

````python
### Data from user ###
north, east = list(map(lambda x: float(x)*np.pi/180, input("Coordinates (north, east): ").split(', ')) )
altitude = float(input("Altitude (in meters): "))
````

````python
### Constants ###
G = 6.67 * 10**-11  # Gravity constant
R = 6_378_100       # Radius of Earth (from Google)
M = 6 * 10**24      # Mass of Earth (from Google)
A = 2_500_000       # Height of atmosphere

V1 = np.sqrt(2*G*M/R)
V2 = np.sqrt(2)*V1
````

````python
### Simple calculations ###
V = np.sqrt(G*M/(R+altitude))
time = 90*60
````

````python
### Useful staff out ###
print(f"{V1=}", f"{V=}", sep='\n')
if V < V1:
    print("Satellite falls :/")
````

````python
### Extra ###
steps = 100  # For animation
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
v, u = np.meshgrid(v, u)
````

````python
### Main code ###
odefun = lambda x, _: \
    np.array([x[3:6], -G*M*x[0:3] / ((np.linalg.norm(x[0:3]))**3)]).reshape(-1)

def get_orbit_n(r, sol_one):
    phi = 90 * np.pi / 180
    z = np.cos(phi)

    p1 = -r[1] / r[0]
    p2 = -z * r[2] / r[0]
    
    a = p1**2 + 1
    b = 2 * p1 * p2
    c = p2**2 - np.sin(phi)**2
    
    if sol_one: y = ((-b + np.sqrt(b**2 - 4 * a * c)) / (2 * a))
    else:       y = ((-b - np.sqrt(b**2 - 4 * a * c)) / (2 * a))

    x = p1 * y + p2

    return np.array([x, y, z])

def main(north, east, velocity):
    init_pos = np.array([
        np.cos(north)*np.cos(east), 
        np.cos(north)*np.sin(east), 
        np.sin(north)
    ])

    n = get_orbit_n(init_pos, False)
    tau = np.cross(n, init_pos)

    global r0 
    r0 = init_pos * (R + altitude)
    v0 = tau * V

    t = np.linspace(0, time, 100_000)  

    x = scipy.integrate.odeint(odefun, list(r0) + list(v0), t, rtol=1e-13, atol=1e-14)
    trajectory = x[:, 0:3]

    trajectory_corrected = np.zeros_like(trajectory)
    for i, current_time in enumerate(t):
        angle_Earth_rotation = -2 * np.pi * current_time / (24 * 60 * 60)
        r = Rotation.from_rotvec(angle_Earth_rotation * np.array([0, 0, 1]))
        current_point = trajectory[i, :]
        current_point_corrected = r.apply(current_point)
        trajectory_corrected[i, :] = current_point_corrected

    return [trajectory_corrected[:, i] for i in range(3)]
````

````python
# def animate(i):
#     velocity = V+(i-5)*1000
#     ax.plot(main(north, east, velocity), 'k', ls='--', linewidth=2)
data = main(north, east, V)
````

````python
### Code for Matplotlib ###
plt.style.use("dark_background")
# fig = plt.figure(figsize=(10, 15), dpi=80)
# ax = fig.add_subplot(projection='3d')
fig, ax = plt.subplots(figsize=(10, 15), dpi=80, subplot_kw={'projection': '3d'})

# Earth
Ex = R * np.cos(u) * np.sin(v)
Ey = R * np.sin(u) * np.sin(v)
Ez = R * np.cos(v)

# anim = mpla.FuncAnimation(plt.gcf(), animate, interval=10)

# ax.plot_surface(Ex, Ey, Ez, alpha=0.2)
ax.plot_wireframe(Ex, Ey, Ez, color="b", alpha=0.2)
ax.plot(*data, 'k', ls='--', linewidth=2)
ax.plot(r0[0], r0[1], r0[2], 'o', color='blue')
ax.plot(0, 0, altitude + R, 'o', color='red')
ax.set_aspect('auto')

plt.show()
````
